# **"Nappy Detective - Pieluszkowy Detektyw"  - aplikacja do wykrywania<br> nieprawidłowego zabarwienia niemowlęcych stolców**


**To innowacyjna aplikacja tworzona z myślą o wczesnym wykrywaniu chorób związanych z nieprawidłowym kolorem stolca niemowląt,<br> takich jak atrezja zewnątrzwątrobowych dróg żółciowych. Dzięki zaawansowanej analizie zdjęć, aplikacja umożliwi szybkie rozpoznanie potencjalnych<br> nieprawidłowości, co może pomóc w wczesnym rozpoczęciu leczenia operacyjnego, tak jak wykazano w badaniach przeprowadzonych w krajach<br> takich jak Tajwan, Japonia czy Kanada.**

**Inspiracją dla aplikacji była inicjatywa PoopMD stworzona przez John Hopkins University, która pomogła znacząco skrócić czas diagnozowania tej choroby.<br> Współwystępowanie żółtaczki i odbarwionych stolców stanowi kluczowy sygnał do rozpoczęcia diagnostyki, a NappyDetective będzie<br> narzędziem wspierającym w tym procesie.**
https://pubmed.ncbi.nlm.nih.gov/26221719/

## Specyfikacja / Specs

* Użytkownik wybiera zdjęcie ze swojego urządzenia i udostępnia aplikacji
* Aplikacja analizuje i klasyfikuje zdjęcie, odrzuca jeżeli jest nieadekwatne
* Aplikacja podaje wynik analizy i doradza dalsze postępowanie
* W przypadkach wątpliwych aplikacja zaleca i umożliwia przesłanie zdjęcia do ośrodka specjalistycznego (email)
* Aplikacja prosi o przesłania zdjęcia do bazy danych w celu zwiększenia skuteczności klasyfikacji
* Aplikacja zawiera treści edukacyjne
* W aplikacji dostępny jest regulamin użytkowania
* Aplikacja umożliwia przekazanie twórcom donacji ("postaw kawę")
* Aplikacja może zawierać reklamy

## Zależności
* Dart 🎯
    * `Flutter` - do budowy aplikacji.
    * `image_picker` - do pobierania zdjęć.
    * `permission_handler` - do uzyskania zgody na dostęp do zdjęć.
    * `device_info_plus` - do sprawdzania wersji urządzenia.
    * `tflite_flutter` - do uruchamiania modeli TensorFlow Lite w aplikacji mobilnej.
    * `image` - do przekształcania zdjęć do modelu.
    * `http` - do przesyłania zdjęć do chmury.
    * `flutter_localizations`, `intl` - do obsługi internacjonalizacji
    
* Python 🐍
    * `pandas` - do przygotowania modeli klasyfikacji.
    * `TensorFlow` - do trenowania i wdrażania modeli uczenia maszynowego.
    * `scikit-learn` - do analizy i przetwarzania danych.
    * `numpy` - do operacji na danych numerycznych.
    * `ResNet` - do trenowania ostatecznego modelu.
